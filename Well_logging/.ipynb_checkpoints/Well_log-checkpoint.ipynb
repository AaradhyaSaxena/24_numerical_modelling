```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_excel('log.xlsx')
data = data[0:9980]
data.fillna(0, inplace=True)
data.head()
```

```python
gr = data['GR (GAPI)']
gr = gr.dropna()
```

1) Calculate V shale

```python
minimum = min(gr)
maximum = max(gr)
print("minimum : ",minimum )
print("maximum : ",maximum )
```

```python
V_shale = (gr-minimum)/(maximum-minimum)
V_shale.head()
```

```python
depth = data['Depth (ft)']
plt.plot(V_shale,depth)
plt.show()
```

# Question_1

```python
# steiber

V_sh_3 = V_shale/(3-2*V_shale)
plt.plot(gr ,V_sh_3)

V_sh_2 = 0.083*(2**((3.7)*(V_shale)) -1)
plt.plot(gr, V_sh_2)

plt.plot(gr,V_shale)
plt.show()
```

```python
gr = gr/150
```

```python
plt.plot(gr,depth)
plt.plot(V_shale,depth)
plt.show()
```

2) Calculate porosity

```python
sonic = data['DTCO (us/f)']
sonic = sonic.dropna()

s_por = (sonic-55.6)/(189-55.6)
s_por = s_por.clip(0,0.6)

plt.plot(V_shale,depth) #  blue => v_shale
plt.plot(s_por,depth)   #  orange => s_por

plt.plot()
plt.show()
```

```python
density = data['RhoB (g/c3)']
density = density.clip(1.6,2.65)

d_por = (2.65 - density)/(2.65 - 1.00)
d_por = d_por.clip(0,0.6)

plt.plot(V_shale,depth)
plt.plot(d_por,depth)
plt.show()
```

```python
n_por = data['NPOR']
```

## Question_2

```python
plt.scatter(V_shale, n_por, alpha =0.2)
plt.scatter(V_shale, s_por, alpha =0.2)
plt.scatter(V_shale, d_por, alpha =0.2)

plt.xlabel("V Shale")
plt.ylabel("Porosity")
plt.show()
```

r_o obtained at 22500 = 1.1585


```python
r_o = 1.1585
depth = 20000
density = 2.3644
d_por = (2.65 - density)/(2.65 - 1)

t_f = 70 + 0.0068*(depth)
t_c = (5/9)*(t_f - 32)
KB = 47.5  ### kelly_bush_depth
pr = 6894.76* 0.45 * (depth - KB) / 10**6

def rho_brine_func(d_por_hp):
    d_por = (2.65 - density)/(2.65 - d_por_hp)
    phi = d_por*d_por
    r_w = (r_o*phi)
    sal = 0.1**((1/0.955*(np.log10(3647.5)))- np.log(r_w*((t_f+6.77)/(81.77)))-0.0123)
    rho_brine = 1 + sal*(0.668 + 0.44*sal + 10**(-6)*(300*pr - 2400*pr*sal + t_c*(80 + 3*t_c - 3300*sal - 13*pr + 47*pr*sal)))
    
    return rho_brine
```

```python
minn = 1
r_calc = 1
loss_arr =[]
x = np.linspace(1,1.2,20001)
for i in x:
    r_br = rho_brine_func(i)
    loss = abs(r_br - i)
    loss_arr.append(loss)
    if(loss< minn):
        minn = loss
        r_calc = i
print("r_calc: ", r_calc)
print("min: ", minn)
```
